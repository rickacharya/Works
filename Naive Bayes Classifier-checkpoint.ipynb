```python
import pandas as pd
from sklearn import datasets
wine_data=datasets.load_wine()
dir(wine_data)
```

```python
df=pd.DataFrame(wine_data.data,columns=wine_data.feature_names)
df.head()
```

```python
wine_data.target_names
```

```python
wine_data.target
```

```python
df['target']=wine_data.target
df[100:140]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(wine_data.data,wine_data.target,test_size=0.2,random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
mn_model=MultinomialNB()
mn_model.fit(X_train,y_train)
```

```python
mn_model.predict(X_test)
```

```python
mn_model.score(X_test,y_test)
```
