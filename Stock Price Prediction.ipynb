```python
df["HL_PCT"]=(df['Adj_High']-df['Adj_Low'])/df['Adj_Close']*100
df['PCT_Change']=(df['Adj_Close']-df['Adj_Open'])/df['Adj_Open']*100
df.head()
```

```python
df=df.drop(['Open', 'High', 'Low', 'Close', 'Volume', 'Dividend', 'Split','Adj_Open', 'Adj_High', 'Adj_Low'],axis=1)
df.head()
```

```python
df=df.fillna(-99999)
df.head()
```

```python
label=df['Adj_Close'].shift(-10)
print(label)
```

```python
import numpy as np
label.dropna(inplace=True)
y=np.array(label)
print(y.shape)
```

```python
X=df[['Adj_Close','Adj_Volume','HL_PCT','PCT_Change']]
X=X[:-10]
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=41)
X_test.head()
```

```python
from sklearn.linear_model import LinearRegression
Lreg=LinearRegression()
Lreg.fit(X_train,y_train)
y_pred=Lreg.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,y_pred)
```
