```python
df_new.corr()
```

```python
df2=df_new.drop(['YEAR','FLIGHT_NUMBER','AIRLINE','DISTANCE','TAIL_NUMBER','TAXI_OUT','SCHEDULED_TIME','DEPARTURE_TIME','WHEELS_OFF','WHEELS_ON','DEPARTURE_TIME','ELAPSED_TIME','AIR_TIME','WHEELS_ON','DAY_OF_WEEK','TAXI_IN','ARRIVAL_TIME','CANCELLATION_REASON'],axis=1)
df2
```

```python
print(df_new.columns.values)
```

```python
df2.fillna(df2.mean(),inplace=True)
```

```python
df2.info()
```

```python

result=[]
for i in df2['ARRIVAL_DELAY']:
    if i>15:
        result.append(1)
    else:
        result.append(0)
print(len(result))        
```

```python
df2['RESULT']=result
```

```python
df2.head()
```

```python
df2.RESULT.value_counts()
```

```python
x=df2.drop(['RESULT','ORIGIN_AIRPORT','DESTINATION_AIRPORT'],axis=1)
x.head()
```

```python
y=df2['RESULT']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_train


```

```python
x_test=scaler.fit_transform(x_test)
x_test
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python

pred=clf.predict_proba(x_test)
pred
```

```python
from sklearn.metrics import roc_auc_score

print(roc_auc_score(pred[:,1],y_test))
```
